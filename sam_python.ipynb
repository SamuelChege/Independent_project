```
import pandas as pd
import numpy as np
from glob import glob
```

```
#loading our datasets

dfc = pd.read_csv('cells_geo.csv', delimiter=';')
df1 = pd.read_csv('/content/Telcom_dataset.csv')
df2 = pd.read_csv('/content/Telcom_dataset2.csv')
df3 = pd.read_csv('/content/Telcom_dataset3.csv')
```

```
#cleaning data
df11 = df1.rename(columns={'SITE_ID' : 'SITE_CODE', 'DATETIME' : 'DATE_TIME', 'PRODUTC' : 'PRODUCT'})
df11.tail(6)
```

```
df22 = df2.rename(columns={'SITE_ID' : 'SITE_CODE'})
df22.head(7)
```

```
df33 = df3.rename(columns={'SIET_ID' : 'SITE_CODE', 'CELLID' : 'CELL_ID'})
df33.tail(7)
```

```
#will be mergign the dataset for three days
tda = pd.concat([df11, df22, df33])
tda.head(7)
```

```
tda.describe()
```

```
#merging all the files into one dataframe df
df = tda.merge(dfc, on='SITE_CODE')
df.tail(7)
```

```
#most used products
#df.groupby(['Animal']).mean()
results = df['VALUE'].groupby(df['VILLES']).sum().sort_values(ascending=False)
results.head(20)
```

```
#business hours start from 8.00AM TO 4.00PM
df.between_time('08:00', '04:00')
```

```
#home hours start from 4pm to 8am
df.between_time('04:01', '07:59')
```

```
#most used city for the three days
result = df.pivot_table(index=['VILLES'], aggfunc='size').sort_values(ascending=False)
result.head(20)
```

```
#the most used products
resultss = df['VALUE'].groupby(df['PRODUCT']).sum().sort_values(ascending=False)
resultss.head(20)
```
